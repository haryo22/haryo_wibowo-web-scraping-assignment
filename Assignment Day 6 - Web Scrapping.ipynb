```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.perpusnas.go.id/'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
list_content = soup.select('.col-xl-4')
```

```python
len(list_content)
```

```python
list_content
```

```python
results = []
for item in list_content:
    data = dict()

    title_element = item.select_one('h3')
    if title_element:
        data['title'] = title_element.text.strip()
    else:
        data['title'] = 'No title found'

    description_element = item.select_one('p')
    if description_element:
        data['description'] = description_element.text.strip()
    else:
        data['description'] = 'No description found'

    link_element = item.select_one('a')
    if link_element and link_element.get('href'):
        data['url'] = link_element.get('href')
    else:
        data['url'] = 'No URL found'

    results.append(data)

df = pd.DataFrame(results)
```

```python
df.to_csv('/content/Assignment Day 6 - Web Scrapping.csv', index=False)

print("CSV file has been saved successfully.")
```
